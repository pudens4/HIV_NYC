```python
import os
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# import data of hiv/aids diagnostic
raw_data = os.path.join('HIV_AIDS_Diagnoses_by_Neighborhood__Sex__and_Race_Ethnicity.csv')
```

```python
#read the data 
read_data = pd.read_csv(raw_data)
read_data.head()
```

```python
read_data.info()
```

```python
read_data.shape
```

```python
# unique values in the columns
read_data.RACE_ETHNICITY.unique()
```

```python
# clean the race ethnicity to convert all to unknown race, 
read_data['RACE_ETHNICITY'] = read_data ['RACE_ETHNICITY'].replace(['All'],['Unknown'])
```

```python
# clean sex column, convert all to unknown
read_data['SEX'] = read_data['SEX'].replace(['All'],['Unknown'])
```

```python
# cleand neighborhood data replacing all with unknown
read_data['Neighborhood'] = read_data['Neighborhood'].replace(['All'],['Unknown'])
```

```python
#count null value
read_data.isnull().sum()
```

```python
# remove null value
data = read_data.dropna().sort_values(by='YEAR')
```
